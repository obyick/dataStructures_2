### Import libraries for data visualization and network analysis

```python
import matplotlib.pyplot as plt  # matplotlib for creating static, animated, and interactive visualizations
import networkx as nx            # networkx to creating, manipulating, and analysing complex networks and graphs
import seaborn as sns            # seaborn for statistical data visualization
```

### Reading `GraphML` file

```python
compose_network = nx.read_graphml('Assets/Graphs/Compose_Network.graphml')
```

### Visualising `In-Degree Distribution`

```python
# Set the plotting style to default
plt.style.use("default")

# Compute and sort in-degree sequence of the network
degree_sequence = sorted([d for n, d in compose_network.in_degree()], reverse=True)

# Create subplots for histograms
fig, ax = plt.subplots(1, 2, figsize=(8, 6))

# Plot histogram of degree sequence with counts
all_data = ax[0].hist(degree_sequence, bins=10)
ax[0].set_title("In-Degree Histogram")
ax[0].set_ylabel("Count")
ax[0].set_xlabel("Degree")

# Plot histogram of degree sequence with probability density
ax[1].hist(degree_sequence, bins=10, density=True)
ax[1].set_title("Probability Density Function")
ax[1].set_ylabel("Probability")
ax[1].set_xlabel("Degree")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

### Advanced `In-Degree Histogram` and `PDF with Dual Axis`

```python
# Set the plotting style to 'fivethirtyeight'
plt.style.use("fivethirtyeight") 

# Create a single subplot with a specified figure size
fig, ax = plt.subplots(1, 1, figsize=(10, 8)) 

# Plot histogram of degree sequence with count labels
sns.histplot(degree_sequence, bins=10, label="Count", ax=ax) 

# Create a secondary y-axis for the probability density function (PDF)
ax2 = ax.twinx()
sns.kdeplot(degree_sequence, color='r', label="Probability Density Function (PDF)", ax=ax2)

# Combine legends from both axes
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

# Remove grids and set axis labels
ax.grid(False)
ax2.grid(False)
ax.set_xlabel("In-Degree")
ax2.set_ylabel("Probability")

# Save the plot as a PNG file with transparent background and high resolution and display the plot
plt.savefig('Assets/Images/Density Function/probability.png', transparent=True, dpi=600, bbox_inches="tight")
plt.show()
```

### Advanced `In-Degree Histogram` and `CDF with Dual Axis`

```python
# Set the plotting style to 'fivethirtyeight'
plt.style.use("fivethirtyeight") 

# Create a single subplot with a specified figure size
fig, ax = plt.subplots(1, 1, figsize=(10, 8)) 

# Plot histogram of degree sequence with count labels
sns.histplot(degree_sequence, bins=10, label="Count", ax=ax) 

# Create a secondary y-axis for the cumulative density function (CDF)
ax2 = ax.twinx()
sns.kdeplot(degree_sequence, color='r', label="Cumulative Density Function (CDF)", ax=ax2, cumulative=True)

# Combine legends from both axes
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

# Remove grids and set axis labels
ax.grid(False)
ax2.grid(False)
ax.set_xlabel("In-Degree")
ax2.set_ylabel("Probability")

# Save the plot as a PNG file with transparent background and high resolution and display the plot
plt.savefig('Assets/Images/Density Function/cumulative.png', transparent=True, dpi=600, bbox_inches="tight")
plt.show()
```
